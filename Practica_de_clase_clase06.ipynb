```python
random_gen = np.random.default_rng()
random_gen
```

```python
lamda = 10
sample_size = 10000

random_poisson_10 = random_gen.poisson(lam=lamda, size = sample_size)
random_poisson_10
```

Ahora graficamos la funcion de densidad de la distribucion de poisson

```python
import seaborn as sns

def distribution_plotter(data, label, bins='auto'):    
    sns.set(rc={"figure.figsize": (7, 5)})
    sns.set_style("white")    
    dist = sns.histplot(data, bins= bins, stat = 'density', kde = False, line_kws={'linewidth':5}, thresh = None)    
    dist.set_title('Distribucion ' + label + '\n', fontsize=16)
    
distribution_plotter(random_poisson_10, "Poisson")
```

```python
lamda = 100
random_poisson_10 = random_gen.poisson(lam=lamda, size = sample_size)
distribution_plotter(random_poisson_10, "Poisson lamda 100", bins='auto')
```

```python
lamda = 1000
random_poisson_10 = random_gen.poisson(lam=lamda, size = sample_size)
distribution_plotter(random_poisson_10, "Poisson lamda 1000", bins='auto')
```

```python
lamda = 10000000000
random_poisson_10 = random_gen.poisson(lam=lamda, size = sample_size)
distribution_plotter(random_poisson_10, "Poisson lamda muy grande", bins='auto')
```
