Observamos algunos registros con valores raros de edad, como 0 años o 99 años, vamos a realizar un filtrado de registros, considerandolos como valores mal cargados.

Dado que no tenemos acceso a la fuente que genero el dataset, utilizaremos nuestro criterio a partir de la distribucion de los datos.

CRITERIO: Optamos por eliminar todos los registros con edades menores a 14 años y mayores a 80 años

```python
df = df[np.logical_and(df.edad > 14, df.edad < 80)]
```

## Variable Target

La pregunta 11 es :

**Después de un episodio violento, ¿se muestra cariñoso/a o atento/a y promete que nunca más lo hará y todo cambiará?**

Siendo las respuestas posibles para todas las preguntas:

A) NUNCA

B) A VECES 

C) SIEMPRE


```python
df.pregunta_11.value_counts(normalize=True)
```

Vamos a tomar esta pregunta como variable target, y consideraremos como caso positivo "1" cuando haya algun tipo de episodio violento (B/C), y 0 en caso de que no haya ocurrido el hecho violento (A)

```python
df['EPISODIO DE VIOLENCIA'] = (np.logical_or(df.pregunta_11=='B',df.pregunta_11=='C')).astype(int)
```

```python
df['EPISODIO DE VIOLENCIA'].value_counts(normalize=True)
```

Podemos observar que nuestra variable target se encuentra un poco desbalanceada, entre ambos casos

# CREACION DEL MODELO DE CLASIFICACION

```python

features_categoricas = ['genero','barrio','pregunta_1','pregunta_2','pregunta_3','pregunta_3','pregunta_4','pregunta_5',	
                        'pregunta_6','pregunta_7','pregunta_8','pregunta_9','pregunta_10','pregunta_12','pregunta_13']

# Aplicamos OneHotEncoding para las variables categoricas

X_categoricas=pd.get_dummies(df[features_categoricas], drop_first=True)

# Separamos el dataframe en las variables explicativas y el vector target

X = pd.concat([df.edad, X_categoricas], axis=1)
y = df['EPISODIO DE VIOLENCIA']

# Creamos el set de testeo y de entrenamiento

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y, random_state = 1)
```

### CREAREMOS UN GRUPO DE VALIDACION Y UN GRUPO DE TESTEO

```python
X_validation, X_test, y_validation, y_test = train_test_split(X_test, y_test, stratify = y_test, random_state = 1)
```

```python
# Estandarizo la variable numerica edad

esc = StandardScaler()
X_train = esc.fit_transform(X_train)
X_validation = esc.transform(X_validation)

# Para el caso de Naive Bayes multinomial, nos pide que las features explicativas sean todas con valores positivos, por eso utilizaremos otro escaler.
esc2 = MinMaxScaler()
X_train_nb = esc2.fit_transform(X_train)
X_validation_nb = esc2.transform(X_validation)

X_test = esc.transform(X_test)
```

### INICIALMENTE INSTANCIAREMOS UN MODELO DE BASELINE

```python
dummy_clf = DummyClassifier(strategy = 'most_frequent', random_state=1)
dummy_clf.fit(X_validation, y_validation)
```

```python
y_pred_dummy_clf = dummy_clf.predict(X_validation)
```

```python
sns.heatmap(confusion_matrix(y_validation, y_pred_dummy_clf), annot= True, fmt='.2f')
plt.show()
```

```python
print(classification_report(y_validation, y_pred_dummy_clf))
```

#### **Conclusion: Con esto podemos inferir que el minimo valor de f1-score de mis modelos, con este dataset, debe ser de almenos un 83%, para la clase positiva**

#### Utilizaremos gridsearch y random search, para encontrar los mejores hiperparametros de los modelos KNN y Regresion Logistica

```python
models = [LogisticRegression(), KNeighborsClassifier()]

params = [
    {'C': [0.00001,0.0001, 0.001, 1],
     'penalty': ['l1', 'l2'],
     'solver': ['saga'],
    'class_weight':[None,'balanced']},
    {'n_neighbors': range(5,50),
     'weights' : ['uniform', 'distance'],
     'p' : [1, 2, 3]}
          ]

folds = StratifiedKFold(n_splits=5 ,shuffle=True, random_state=1)
```

Evaluamos la mejor configuración de hiperparámetros de cada modelo: LogReg y KNN

Debido a que a nosotros nos interesa identificar cuando es un noviazgo violento, queremos darle mas importancia a los T.P. , con lo cual, vamos a utilizar como metrica para comparar clasificadores el RECALL, o TRUE POSITIVE RATE

```python
gds = []
for i in range(len(models)):
    gs = GridSearchCV(estimator=models[i], param_grid=params[i], scoring='f1', cv=folds, n_jobs=8)
    print (gs)
    fit = gs.fit(X_train, y_train)
    gds.append(fit)
```

```python
for i in gds:
    print (i.best_score_)
    print (i.best_estimator_)
    print (i.best_params_)
    print()
```

## MODELO KNN

```python
gds[1].best_estimator_
```

```python
knn = gds[1].best_estimator_

knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_validation)

recall_score(y_validation, y_pred_knn)
```

```python
print(classification_report(y_validation, y_pred_knn))
```

```python
sns.heatmap(confusion_matrix(y_validation, y_pred_knn, normalize= 'true'), annot= True,fmt='.2f')
plt.xlabel('PREDICCIONES DEL MODELO')
plt.ylabel('DATOS REALES')
plt.title('MATRIZ DE CONFUSION KNN VALIDATION SET')
plt.show()
```

Vemos que el modelo knn, tiene una buena clasificacion para la clase positiva, pero una baja clasificacion para la clase negativa, con casi un 70% de FALSOS POSITIVOS

### CURVA ROC Y AUC

```python
y_pred_proba = knn.predict_proba(X_validation)
y_pred_proba
```

```python
# Curva ROE de validation
fpr,tpr,th = roc_curve(y_validation, y_pred_proba[:,1])

df_plot = {'FALSE POSITIVE RATE':fpr,
           'TRUE POSITIVE RATE':tpr,
           'Threshold':th}

# Curva ROE de train
y_pred_knn_train = knn.predict_proba(X_train)
fpr,tpr,th = roc_curve(y_train, y_pred_knn_train[:,1])

df_toplot = pd.DataFrame({'fpr':fpr,
                          'tpr':tpr,
                          'th':th})

f,(axis1,axis2) = plt.subplots(1,2,figsize=(15,5))

axis1.plot(df_plot['FALSE POSITIVE RATE'], df_plot['TRUE POSITIVE RATE'])
axis2.plot(df_toplot['fpr'], df_toplot['tpr'])
axis1.set_xlabel('FALSE POSITIVE RATE')
axis1.set_ylabel('TRUE POSITIVE RATE')
axis1.set_title('Curva ROE KNN VALIDATION')

axis2.set_xlabel('FALSE POSITIVE RATE')
axis2.set_ylabel('TRUE POSITIVE RATE')
axis2.set_title('Curva ROE KNN TRAIN')

axis1.grid()
axis2.grid()

plt.show()
```

Observamos que no hay overfitting, ya que el modelo performa de manera parecida tanto con el set de entrenamiento como de testeo

**CALCULAMOS EL AUC DEL MODELO KNN PARA EL SET DE VALIDATION**

```python
AUC_knn = auc(df_plot['FALSE POSITIVE RATE'], df_plot['TRUE POSITIVE RATE'])
AUC_knn
```

## MODELO LOGISTIC REGRESSION

```python
lr = gds[0].best_estimator_

lr.fit(X_train, y_train)

y_pred_lr = lr.predict(X_validation)

recall_score(y_validation, y_pred_lr)
```

```python
pd.Series(y_pred_lr).value_counts()
```

```python
print(classification_report(y_validation, y_pred_lr))
```

```python
sns.heatmap(confusion_matrix(y_validation, y_pred_lr, normalize= 'true'), annot= True,fmt='.3g')
plt.xlabel('PREDICCIONES DEL MODELO')
plt.ylabel('DATOS REALES')
plt.title('MATRIZ DE CONFUSION REGRESION LOGISTICA VALIDATION SET')
plt.show()
```

Podemos observar, que tiene un excelente desempeño para classificar a la clase positiva, pero un bajo desempeño para la clase negativa. Tambien obsevamos una gran cantidad de Falsos Positivos, pero nosotros, lo que nos da cuenta de nuestro valor de RECALL para la clase positiva de 0.94. Este modelo nos sirve, ya que preferimos sacrificar algunos FP, antes que tener algun FN

### CURVA ROC Y AUC

```python
y_pred_lr_test = lr.predict_proba(X_validation)

# Curva ROE para el set de validation
fpr,tpr,th = roc_curve(y_validation, y_pred_lr_test[:,1])

df_toplot_validation_lr = pd.DataFrame({'fpr':fpr,
                          'tpr':tpr,
                          'th':th})

# Curva ROE para el set de train
y_pred_lr_train = lr.predict_proba(X_train)
fpr,tpr,th = roc_curve(y_train, y_pred_lr_train[:,1])

df_toplot_train_lr = pd.DataFrame({'fpr':fpr,
                          'tpr':tpr,
                          'th':th})

f,(ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))
ax1.plot(df_toplot_validation_lr['fpr'], df_toplot_validation_lr['tpr'])
ax1.set_xlabel('FALSE POSITIVE RATE')
ax1.set_ylabel('TRUE POSITIVE RATE')
ax1.set_title('Curva ROE Regresion Logistica VALIDATION')
ax1.grid()

ax2.plot(df_toplot_train_lr['fpr'], df_toplot_train_lr.tpr)
ax2.set_xlabel('FALSE POSITIVE RATE')
ax2.set_ylabel('TRUE POSITIVE RATE')
ax2.set_title('Curva ROE Regresion Logistica TRAIN')
ax2.grid()

plt.show()
```

Observamos que no overfittea, ya que tiene una performance parecia tanto con el set de train, como con el de validation

**CALCULAMOS EL AUC PARA EL MODELO DE REGRESION LOGISTICA CON EL SET DE VALIDACION**

```python
AUC_lr_validation = auc(df_toplot['fpr'], df_toplot['tpr'])
AUC_lr_validation
```
