## Matplotlib
Dado las variables randomio1 y randomio2, generar:

Un diagrama de líneas que grafique ambas series de datos. Una linea debe ser roja y con tipo de línea punteada, la otra verde con linea continua. Agregar una leyenda en la parte inferior izquierda que describa cada serie.

Un diagrama de dispersión que grafique ambas series de datos. Los puntos deben ser de color azul. Agregar titulo y nombre de los ejes x e y.

Un diagrama de barras de las dos variables, pero solo para los 10 primeros elementos. El ancho de las barras desde ser de 0.25. Las barras deben ser negra y verde.

```python
import pandas as pd
import numpy.random as rd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rd.seed(0)
randomio1 = rd.randint(20, size=30)
randomio2 = rd.randint(20, size=30)
```

```python
randomio1
```

```python
fig = plt.figure(figsize=(6, 3))
ax = plt.axes()

ax.plot(randomio1, color ='r', linestyle=':', label = 'Randomio1')
ax.plot(randomio2, color= 'g', linestyle='-', label = 'Randomio2')

fig.suptitle("TITULO FIGURA")
ax.set(xlabel= "TITULO EJE X", ylabel= 'TITULO EJE Y', title="TITULO PLOT")



ax.legend(loc='lower left', title = 'Series')
```

```python
fig = plt.figure(figsize=(12, 6))
ax1 = plt.axes()
ax1.grid()
ax1.scatter(randomio1,randomio2, color='b')
ax1.set(xlabel= 'randomio 1', ylabel = 'randomio2', title="Scatter Plot")
```
