# 2. Preprocesando cada grupo de columnas

Observamos que el preprocesamiento de los datos requiere distintos enfoques para distintos tipos de columnas: algunas requieren imputación, otras requieren generar variables Dummies y otras sería conveniente estandarizarlas.

### 2.b Transformador de Edad

Dependiendo la estrategia que hayamos decido vamos a necesitar imputar los datos de edad faltantes, ya sea usando un transformador del módulo de pre-procesamiento o crear un transformador custom transformer.
Esto podría implicar:

- Seleccionar una o más columnas
- Llenar los datos faltantes
- Escalar los valores de Edad

```python
from sklearn.pipeline import make_pipeline, make_union, Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.base import BaseEstimator, TransformerMixin
```

```python
class Feature_selector(BaseEstimator, TransformerMixin):
    
    def __init__(self, feature_selector):
        self.feature_selector = feature_selector
    
    def fit(self, X, y= None):
        return self
    
    def transform(self, X, y=None):
        return pd.DataFrame(X[self.feature_selector])
```

```python
fs = Feature_selector('Age')
```

```python
fs.transform(df).head()
```

```python
steps = [('feature_selection', Feature_selector('Age')),
         ('imputacion', SimpleImputer(strategy = 'mean')),
         ('estandarizacion', StandardScaler())]

pipe_age = Pipeline(steps)


pipe_age.fit_transform(df)
```

## 3. Variables Categóricas

"Embarked" y "Pclass" son variables categóricas. Usá la función [`get_dummies`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html) de pandas para crear columnas correspondientes a los valores de las mismas.

"Embarked" tiene un par de datos faltantes. Llenalos con el puerto de embarque más común en el dataset.

Sugerencia: Crear un transformador custom que "envuelva" el uso de get_dummies.

```python
df.Embarked.value_counts()
```

```python
df.Pclass.value_counts()
```

```python
# imputo los faltantes


df.Embarked = df.Embarked.fillna('S')
```

```python
class Category_transformer(BaseEstimator, TransformerMixin):
    
    def __init__(self, features_categoricas):
        self.features_categoricas = features_categoricas
        
    def fit(self, X, y=None):
        return self
    
    def transform(self, X, y=None):
        return pd.get_dummies(X[self.features_categoricas], columns = self.features_categoricas)

```

```python
features_categoricas = ['Embarked', 'Pclass']

generame_dummies = Category_transformer(features_categoricas) # Aca lo instancie
```

```python
generame_dummies.fit_transform(df).head()
```

```python
step_make_dummies = [('make_dummie', Category_transformer(features_categoricas))]

pipe_make_dummies = Pipeline(step_make_dummies)
```

```python
pipe_make_dummies.fit_transform(df)
```

## 4. Tarifa

Escalar el atributo "Fare" (Tarifa) usando uno de los escaladores existentes en el módulo de preprocesamiento. 

```python
steps_2 = [('seleccion', Feature_selector('Fare')),
           ('escalado', StandardScaler())]

pipe_2 = Pipeline(steps_2)
```

```python
pipe_2.fit_transform(df)
```

## 5. Otras columnas

```python
otras_columnas=['SibSp', 'Parch']

step_3 =  [('seleccion', Feature_selector(otras_columnas)),
           ('escalado', StandardScaler())]

pipe_3 = Pipeline(step_3)
```

```python
pipe_3.fit_transform(df)
```

## 6. Union

Utilizá una FeatureUnion o la función make_union para combinar todos los pipelines que has creado.

```python

union=make_union( pipe_age,
                  pipe_make_dummies,
                  pipe_2,
                  pipe_3 )


union.fit_transform(df)
```

## 7. Predicciones

Ahora utilicemos GridSearch para evaluar la performance de estas transformaciones, seguidas de un modelo de regresión logística. 
Para esto exploren distintos valores de parámetros para C.


Pimero separar las X y las y

```python
X = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
y = df['Survived']
X.head()
```

Separar los datos de train y test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1203, stratify = y)
```

Aplicar el `pipeline` con todos los objetos transformadores del dataframe

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
final_steps = [('data_engeneering',union),
               ('estimator', LogisticRegression())]

final_pipe = Pipeline(final_steps)
```

```python
param_grid = [{'estimator__C':np.linspace(0.01, 1000, 1000)},
             {'estimator__penalty':['l1','l2', None]}]

skf = StratifiedKFold(n_splits=5, shuffle = True, random_state = 1203)


gs = GridSearchCV(final_pipe, param_grid, n_jobs=-1, cv=skf, scoring = 'f1_macro') # f1_macro en datasets unbalanceds penaliza mas a equivocarse en la clase minoritaria
```

```python
gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

```python
gs.best_estimator_
```

```python
gs.best_score_
```

#### Performance sobre datos nuevos



```python
best_estimator=gs.best_estimator_
```

```python
print(classification_report(y_test, best_estimator.predict(X_test)))
print(accuracy_score(y_test, best_estimator.predict(X_test)))
```
