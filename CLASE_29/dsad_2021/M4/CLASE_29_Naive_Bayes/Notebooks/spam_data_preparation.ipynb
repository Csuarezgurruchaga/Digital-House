```python
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import string
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import LabelBinarizer
```

```python
df = pd.read_csv('../Data/spam.csv', encoding='iso8859_14')
df
```

```python
df.drop(labels=df.columns[2:],axis=1,inplace=True)
df.columns=['target','text']
```

```python
cv = CountVectorizer()
X = cv.fit_transform(df['text']).toarray()

y = df['target']
print(X.shape,y.shape)
```

```python
X
```

```python
Y = pd.DataFrame(y, columns=['target'])
Y
```

```python
X.shape
```

```python
result = pd.DataFrame(X, columns = cv.get_feature_names())
to_write = pd.concat([result, Y], axis=1)
to_write.shape
```

```python
to_write.to_csv('../Data/spam_count_vectorizer.csv', sep='\t', index=False)
```
